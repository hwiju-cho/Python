### Recurrence

```python
def fac(n):
    if n==1:
        return 1
    return fac(n-1)*n
```

```python
print(fac(4))
```

```python
def fac(n):
    if n==1:
        print("fac(1) => return 1")
        return 1
    print("fac({0}) => fac({1}) * {2}".format(n,n-1,n))
    return fac(n-1)*n
```

```python
fac(4)
```
